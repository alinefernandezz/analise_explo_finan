# Análise Exploratória de Dados (EDA) sobre Vendas de Produto Financeiro

Nesta análise, realizei uma Análise Exploratória de Dados (EDA) em um conjunto de dados contendo informações sobre as vendas de um produto financeiro. O conjunto de dados inclui variáveis como idade, profissão, estado civil, nível de escolaridade e se o cliente comprou ou não o produto financeiro.

```python
#Importando Bibliotecas Necessárias
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Definindo o tema visual
sns.set(style="whitegrid")
```

```python
#Passo 2: Carregando e Inspecionando os Dados
```

```python
# Carregar o conjunto de dados
df = pd.read_excel('bank_marketing_ficticio.xlsx')

# Informações básicas sobre o conjunto de dados
print(df.info())
print(df.describe())
print(df.head())
```

Este passo nos permite entender a estrutura dos dados, incluindo os tipos de variáveis, valores ausentes e estatísticas básicas.

```python
#Passo 3: Analisando a Distribuição das Principais Variáveis
```

```python
# Distribuição da Idade
plt.figure(figsize=(10, 6))
sns.histplot(df['Idade'], bins=30, kde=True, color='blue')
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Compreender a distribuição de idades nos ajuda a identificar os principais grupos demográficos que são alvo do produto financeiro.

```python
#Passo 4: Analisando Variáveis Categóricas
```

```python
# Gráfico de barras para Profissão
plt.figure(figsize=(12, 6))
sns.countplot(y='Profissão', data=df, order=df['Profissão'].value_counts().index, palette='coolwarm')
plt.title('Distribuição das Profissões')
plt.xlabel('Contagem')
plt.ylabel('Profissão')
plt.show()

# Gráfico de pizza para Estado Civil
df['Estado_Civil'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("coolwarm", 3), startangle=140, figsize=(8, 8))
plt.title('Distribuição do Estado Civil')
plt.ylabel('')
plt.show()
```

Essas visualizações fornecem insights sobre as profissões e estados civis mais comuns entre a base de clientes.

```python
#Passo 5: Análise de Correlação
```

```python
# Matriz de correlação
corr_matrix = df.corr(numeric_only=True) # Include numeric_only=True
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação')
plt.show()
```

Este mapa de calor nos ajuda a identificar relações entre diferentes variáveis numéricas, como a correlação entre saldo em conta corrente e a compra do produto.

```python
#Passo 6: Analisando as Taxas de Compra do Produto
```

```python
# Porcentagem de clientes que compraram o produto
purchase_rate = df['Cliente_Comprou_o_Titulo?'].value_counts(normalize=True) * 100
print(purchase_rate)

# Plotando a taxa de compra
sns.barplot(x=purchase_rate.index, y=purchase_rate.values, palette='coolwarm')
plt.title('Taxa de Compra do Produto Financeiro')
plt.xlabel('Comprou')
plt.ylabel('Porcentagem')
plt.show()
```

Compreender a taxa de compra é crucial para avaliar a eficácia das estratégias de marketing.

```python
#Passo 7: Conclusão
```

Esta análise proporcionou uma visão abrangente do conjunto de dados. Os principais insights incluem a distribuição de idades, profissões e estados civis, bem como as correlações entre vários indicadores financeiros. Esses insights podem ajudar a direcionar segmentos específicos de clientes e melhorar as estratégias de marketing.
